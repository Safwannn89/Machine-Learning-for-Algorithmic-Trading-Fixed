```python
import pandas_datareader.data as web
import matplotlib.pyplot as plt
import datetime
```

```python
import seaborn as sns

# Now set the visual style
sns.set_style('whitegrid')
```

```python
def plot_gdp():
    start = datetime.datetime(2000, 1, 1)
    end = datetime.datetime.today()
    gdp = web.DataReader('GDP', 'fred', start, end)
    gdp.plot(title="US GDP Over Time")
    plt.ylabel("Billions of Dollars")
    plt.show()

# Call the function
plot_gdp()
```

```python
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
import seaborn as sns

# Define date range
start = datetime.datetime(2000, 1, 1)
end = datetime.datetime.today()

# Get Crude Oil Prices (WTI)
oil = web.DataReader('DCOILWTICO', 'fred', start, end)

# Plot (styled + only once)
ax = oil.plot(lw=2, title='WTI Crude Oil Price', figsize=(12, 4))
sns.despine()
plt.ylabel("Dollars per Barrel")
plt.show()

```
